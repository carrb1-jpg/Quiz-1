```python
import numpy as np
import pandas as pd

# Method 1: IQR Outlier Detection
def detect_outliers_iqr(data, column, factor=1.5):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - factor * IQR
    upper = Q3 + factor * IQR
    return (data[column] < lower) | (data[column] > upper)

# Method 2: Z-Score Outlier Detection
def detect_outliers_zscore(data, column, threshold=3):
    mean = data[column].mean()
    std = data[column].std()
    z_scores = (data[column] - mean) / std
    return np.abs(z_scores) > threshold

# Apply both methods to all features + target
outlier_summary = {}
for col in feature_cols + [target_col]:
    iqr_outliers = detect_outliers_iqr(df, col).sum()
    z_outliers = detect_outliers_zscore(df, col).sum()
    outlier_summary[col] = {"IQR": iqr_outliers, "Z-score": z_outliers}

outlier_df = pd.DataFrame(outlier_summary).T
print("Outlier counts per feature:")
display(outlier_df)
```

```python
plt.figure(figsize=(12,8))
df[feature_cols + [target_col]].boxplot()
plt.title("Boxplots Before Outlier Removal")
plt.xticks(rotation=45)
plt.show()

# Remove outliers using IQR (example choice)
clean_df = df.copy()
for col in feature_cols + [target_col]:
    mask = detect_outliers_iqr(clean_df, col)
    clean_df = clean_df[~mask]

print("Original shape:", df.shape)
print("After removing outliers (IQR):", clean_df.shape)

# Boxplots after outlier removal
plt.figure(figsize=(12,8))
clean_df[feature_cols + [target_col]].boxplot()
plt.title("Boxplots After Outlier Removal (IQR)")
plt.xticks(rotation=45)
plt.show()
```

The IQR method detected many more outliers compared to the Z-score method, which makes sense because several features are skewed.
Boxplots before removal showed extreme values in features such as average rooms, population, and target house value.
After applying the IQR method, these extremes were reduced, and the dataset size dropped from 20,640 to about 16,322 rows.
I chose the IQR method because it is more robust to skewed data, which is common in housing datasets.

```python
import numpy as np

def euclidean_distance(point1, point2):
    """
    Calculate Euclidean distance between two points
    """
    return np.sqrt(np.sum((point1 - point2) ** 2))

def manhattan_distance(point1, point2):
    """
    Calculate Manhattan distance between two points
    """
    return np.sum(np.abs(point1 - point2))

def minkowski_distance(point1, point2, p=2):
    """
    Calculate Minkowski distance between two points
    (p=1 => Manhattan, p=2 => Euclidean)
    """
    return np.sum(np.abs(point1 - point2) ** p) ** (1/p)

# Quick test
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Euclidean:", euclidean_distance(a, b))
print("Manhattan:", manhattan_distance(a, b))
print("Minkowski (p=3):", minkowski_distance(a, b, p=3))
```

```python
import numpy as np

class CustomKNN:
    def __init__(self, k=5, distance_metric='euclidean', weights='uniform'):
        """
        Custom implementation of k-NN
        k: number of neighbors (int)
        distance_metric: 'euclidean', 'manhattan', 'minkowski'
        weights: 'uniform' or 'distance'
        """
        self.k = int(k)
        self.distance_metric = distance_metric
        self.weights = weights
        self.X_train = None
        self.y_train = None

    def fit(self, X, y):
        """Store training data"""
        self.X_train = np.asarray(X, dtype=float)
        self.y_train = np.asarray(y, dtype=float)
        if self.X_train.ndim != 2:
            raise ValueError("X must be 2D array.")
        if self.y_train.ndim != 1:
            raise ValueError("y must be 1D array.")
        if len(self.X_train) != len(self.y_train):
            raise ValueError("X and y must have same length.")

    def _calculate_distance(self, p1, p2):
        """Calculate distance using the selected metric"""
        if self.distance_metric == 'euclidean':
            return np.sqrt(np.sum((p1 - p2) ** 2))
        elif self.distance_metric == 'manhattan':
            return np.sum(np.abs(p1 - p2))
        elif self.distance_metric == 'minkowski':
            return np.sum(np.abs(p1 - p2) ** 3) ** (1/3)  # p=3
        else:
            raise ValueError("Unsupported distance metric.")

    def _get_neighbors(self, x):
        """Return a list of (index, distance) for the k nearest neighbors"""
        dists = [(i, self._calculate_distance(x, xi)) for i, xi in enumerate(self.X_train)]
        dists.sort(key=lambda t: t[1])
        return dists[:self.k]

    def predict_single(self, x):
        """Predict target for a single sample (uniform or distance weights)"""
        neighbors = self._get_neighbors(np.asarray(x, dtype=float))
        idxs = [i for i, _ in neighbors]
        dvals = np.array([d for _, d in neighbors], dtype=float)
        yvals = self.y_train[idxs]

        if self.weights == 'uniform':
            return float(np.mean(yvals))
        elif self.weights == 'distance':
            w = 1.0 / (dvals + 1e-5)  # avoid div-by-zero
            return float(np.average(yvals, weights=w))
        else:
            raise ValueError("weights must be 'uniform' or 'distance'.")

    def predict(self, X):
        """Predict targets for multiple samples"""
        X = np.asarray(X, dtype=float)
        return np.array([self.predict_single(row) for row in X])

    def score(self, X, y):
        """R^2 score on given test data"""
        y = np.asarray(y, dtype=float)
        y_pred = self.predict(X)
        ss_res = np.sum((y - y_pred) ** 2)
        ss_tot = np.sum((y - np.mean(y)) ** 2)
        return 1 - ss_res / ss_tot
```
